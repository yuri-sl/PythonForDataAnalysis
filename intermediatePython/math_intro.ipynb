loc and iloc (3)

It's also possible to select only columns with loc and iloc. In both cases, you simply put a slice going from beginning to end in front of the comma:

cars.loc[:, 'country']
cars.iloc[:, 1]

cars.loc[:, ['country','drives_right']]
cars.iloc[:, [1, 2]]


There was an error in your code

The error comes from the following part of the code:

print(cars.loc[[:,"drives_right"]])

What is the issue?

The problem here is with how the .loc indexer is being used. In pandas, .loc expects the first argument to specify the row labels (or a slice of them), and the second argument to specify the column labels. When you write [ : , "drives_right" ] inside a single set of square brackets, Python interprets this as trying to create a list with a slice and a string, which is not valid syntax and leads to a SyntaxError.

To select columns using .loc, you should separate the row and column selectors with a comma inside the brackets, not wrap them in a list. If you want to select all rows for a specific column as a DataFrame, you should use:

cars.loc[:, ["drives_right"]]

Notice that the row selector (:) and the column selector (["drives_right"]) are separated by a comma, and the column selector is a list to ensure the result is a DataFrame.

How do you fix it?

Replace the incorrect line with the following:

print(cars.loc[:, ["drives_right"]])

This will correctly print the drives_right column as a DataFrame. You're on the right track—just remember that .loc uses a comma to separate row and column selections, not a list.

```python

# Import cars data
import pandas as pd
cars = pd.read_csv('cars.csv', index_col = 0)

# Print out drives_right column as Series
print(cars.loc[:,"drives_right"])

# Print out drives_right column as DataFrame
print(cars.loc[:,["drives_right"]])

# Print out cars_per_cap and drives_right as DataFrame
print(cars.loc[:,["cars_per_cap","drives_right"]])
```
